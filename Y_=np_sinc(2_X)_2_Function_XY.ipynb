```
%tensorflow_version 1.x 
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
plt.style.use('classic')
sess = tf.InteractiveSession()
```

```

X =np.expand_dims(np.arange(0.0, 2.0, 0.01),1)
Y =np.sinc(2*X)/2

x = tf.placeholder(tf.float64, [200,1], name='x')
y = tf.placeholder(tf.float64, [200,1], name='y')
```

```
#Model_1
input_layer = tf.layers.dense(x, 1)
hidden_layer1 = tf.layers.dropout(input_layer,0.25)
hidden_layer2 = tf.layers.dense(hidden_layer1,90,activation=tf.nn.relu)
output_layer = tf.layers.dense(hidden_layer2,1)

Loss =tf.losses.mean_squared_error(y , output_layer)
Optimizer = tf.train.AdamOptimizer(learning_rate= 0.001).minimize(Loss)
init = tf.global_variables_initializer()
```

```
#model_1
loss_list=[]
sess.run(init)
for i in range(0,10000):
  fd ={x:X, y:Y}
  _, loss_val = sess.run([Optimizer, Loss], feed_dict=fd)
  #print ('loss = %s' % loss_val)
  loss_list.append(loss_val)

YP1 = sess.run(output_layer,feed_dict={x:X})
```

```
#model_1
# Plot training  loss values
import matplotlib.pyplot as plt
plt.plot(loss_list)
plt.title('Model_1 loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```

plt.plot(X,YP1, label='Model 1')

plt.plot(X,Y, label='y=sinc(2*X)/2')
plt.title("Function Models")
plt.xlabel('x')
plt.ylabel('y')
#plt.show()
plt.legend(framealpha=1, frameon=True)
```
